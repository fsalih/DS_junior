## Разделение выборки

```python
train, test = train_test_split(df, test_size=0.2, random_state=RS)
```

## Работа с фичами

```python
num_scaler = StandardScaler()
df_num_scaled = num_scaler.fit_transform(df[numeric_cols])
df_num_scaled = pd.DataFrame(data=df_num_scaled, columns=num_scaler.get_feature_names_out())
df_num_scaled
```

```python
ohe = OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore')
df_cat_transformed = pd.DataFrame(data=ohe.fit_transform(df[categorical_cols]), columns=ohe.get_feature_names_out())
df_cat_transformed
```

```python
df_prepared = pd.concat([df_num_scaled, df_cat_transformed], axis=1)
df_prepared
```

```python
model = SVC(class_weight='balanced', random_state=RS, verbose=100, probability=True)
```

```python
%%time
model.fit(df_prepared, df[target])
```

```python
probe_train = model.predict_proba(df_prepared)
```

```python
pred_train = model.predict(df_prepared)
```

```python
confusion_matrix(df[target], pred_train)
```
