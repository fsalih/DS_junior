```python
X = np.array([1, 3, 4, 5, 7])
X = np.reshape(X, (5,1))
X
```

```python
Y = np.array([6, 16, 21, 26, 36])
Y = np.reshape(Y, (5,1))
Y
```

```python
E = np.ones((5, 1))
E
```

```python
X = np.hstack([E, X])
X
```

```python
X.T
```

```python
A = np.dot(X.T, X)
A
```

```python
np.linalg.det(A)
```

```python
A_inv = np.linalg.inv(A)
A_inv
```

```python
B = np.dot(X.T, Y)
B
```

```python
w = np.dot(A_inv, B)
w
```

```python
X
```

```python
w
```

```python
np.dot(X, w)
```

```python
np.dot(X, w) == Y
```
