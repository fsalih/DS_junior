### 1. Импорт библиотек

```python
# Ваш код здесь
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error, mean_squared_error
```

### 2. Загрузка данных

```python
# Ваш код здесь
df = pd.read_csv('4.7 sales.csv')
df
```

### 3. Первичный анализ

1. Проведите анализ данных, посмотрите на индивидуальные графики зависимости целевой функции и отдельной переменной. Нередко, если признаков в выборке не так много, удобно построить [sns.pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html).
2. Закодируйте категориальные признаки (например, с помощью [pd.get_dummies](https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html)).
3. Подумайте, можно ли здесь создать дополнительные переменные из имеющихся.
4. Сделайте предварительные выводы.

```python
# Ваш код здесь
df.Street.value_counts()
```

```python
df.SaleCondition.value_counts()
```

```python
numeric_features = ['SalePrice', 'GrLivArea', 'GarageArea', 'OverallQual']
```

```python
sns.pairplot(df[numeric_features])
```

```python
df[numeric_features].corr()
```

```python
df.Street = pd.get_dummies(df.Street).astype('int')['Grvl']
```

```python
sc_df = pd.get_dummies(df.SaleCondition, drop_first=True, dtype='int')
sc_df
```

```python
df = pd.concat([df, sc_df], axis=1)
df
```

Зависимость SalePrice от OverallQuality похожа на квадратичную. Можно попробовать сделать из нее новую фичу.
Между другими параметрами нет однозначной зависимости. 

```python
df['OverallQual2'] = df['OverallQual']**2
```

### 4. Разделение на обучающую и тестовую выборки

```python
# Ваш код здесь
train, test = train_test_split(df.drop(columns=['SaleCondition']))
```

### 5. Обучение модели линейной регрессии

Дополнительно можно взглянуть на коэффициенты и смещение обученной модели.

```python
# Ваш код здесь
# Можно масштабировать данные, чтобы понять что больше влияет на цену, смысл масштабирования понятен 
features = ['GrLivArea', 'GarageArea', 'OverallQual', 'Street', 'AdjLand', 'Alloca', 'Family', 'Normal', 'Partial', 
            'OverallQual2']
target = 'SalePrice'
model = LinearRegression()
model.fit(train[features], train[target])
print('Коэффициенты модели:', model.coef_)
print('Смещение модели:', model.intercept_)
```

### 6. Получение предсказаний для обучающей и тестовой выборок

```python
# Ваш код здесь
pred_train = model.predict(train[features])
pred_test = model.predict(test[features])
```

### 7. Проверка предсказаний

1. Проверьте качество модели на обучающей и тестовой выборках с помощью MAE, MSE.
2. Постройте диаграмму рассеяния целевой и предсказанной переменных. 

```python
# Ваш код здесь
print('MAE:\n', 'train:', mean_absolute_error(train[target], pred_train))
print(' test:', mean_absolute_error(test[target], pred_test))

print('MSE:\n', 'train:', mean_squared_error(train[target], pred_train, squared=False))
print(' test:', mean_squared_error(test[target], pred_test, squared=False))
```

```python
plt.figure(figsize=(8,8))
plt.scatter(train[target], pred_train, s=4, label='train')
plt.scatter(test[target], pred_test, s=4, label='test', alpha=0.7)
plt.xlabel('true')
plt.ylabel('pred')
plt.legend()
plt.xlim(0, 500_000)
plt.ylim(0, 500_000)
plt.plot([0, 500_000], [0, 500_000], c='red')
plt.show()
```

В моем понимании качество модели низкое. 
Умею улучшать линейные модели возведением признаков в степень, логарифмированием и т.д.

### Дополнительно

```python
from sklearn.metrics import r2_score
print('r2_score:\n', 'train:', r2_score(train[target], pred_train))
print(' test:', r2_score(test[target], pred_test))
```
